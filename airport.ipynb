```python
from scipy import stats
z_scores = stats.zscore(df.select_dtypes(include=['float64', 'int64']))
df_outliers = df[(z_scores > 3).any(axis=1) | (z_scores < -3).any(axis=1)]
print(df_outliers.shape)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['fare'], kde=30, color='blue')
plt.title('Distribution of fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['nsmiles'], kde=True, color='green')
plt.title('Distribution of Miles')
plt.xlabel('Miles')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='nsmiles', y='fare', data=df)
plt.title('Relationship between Fare and Miles')
plt.xlabel('Miles')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='fare', data=df.groupby('Year')['fare'].mean().reset_index(), marker='o')
plt.title('Average Fare by Year')
plt.xlabel('Year')
plt.ylabel('Average Fare')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_cities = df.groupby('city1')['passengers'].sum().nlargest(10).reset_index()
sns.barplot(x='passengers', y='city1', data=top_cities)
plt.title('Top 10 cities by passanger Count')
plt.xlabel('Total Passangers')
plt.ylabel('City 1')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='carrier_lg', y='large_ms', data=df)
plt.title('Distribution of large Carrier Market Share')
plt.xlabel('corrier')
plt.xticks(rotation=90)
plt.ylabel('Market Shape')
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='carrier_low', y='lf_ms', data=df)
plt.title('Distribution of low carrier Market Share')
plt.xlabel('Carrier')
plt.xticks(rotation=90)
plt.ylabel('Market Share')
plt.show()
```

```python
plt.figure(figsize=(20, 8))
sns.boxplot(x='quarter', y='passengers', hue='Year', data=df)
plt.title('Passenger Count Variation by Quarter and Year')
plt.xlabel('Quarter')
plt.ylabel('Passenger Count')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderax)
```
